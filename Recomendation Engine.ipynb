# Data Preprocessing

```python
# merge the two dataset 
movies = movies.merge(credit,on='title')
```

```python
# Feature Selection For the recomendation engine
movies = movies[["genres","id","overview","title","cast","crew"]]
```

```python
#finding Null values
movies.isnull().sum()
```

```python
# Drop the Null values
movies.dropna(inplace=True)
```

```python
# For the string conversion
import ast
```

```python
# Fetching the Genres of Movies
def convert(text):
    list1=[]
    for i in ast.literal_eval(text):
        list1.append(i['name'])
    return list1
```

```python
movies["genres"]=movies['genres'].apply(convert)
```

```python
movies.sample(1)
```

```python
movies['cast'][0]
```

```python
# Finding  Lead Cast
def top_cast(text):
    cast=[]
    count=0
    for i in ast.literal_eval(text):
        if count!=2:
            cast.append(i['name'])
            count+=1
    return cast
```

```python
movies['cast']=movies['cast'].apply(top_cast)
```

```python
movies['crew'][0]
```

```python
# Fetching The Director
def dire(text):
    director = []
    for i in ast.literal_eval(text):
        if i['job']=="Director":
            director.append(i['name'])
    return director
```

```python
movies['crew']=movies['crew'].apply(dire)
```

```python
movies['overview'][0]
```

```python
# Excluding the spaces between the text
def replace(text):
    l= []
    for i in text:
        l.append(i.replace(" ",""))
    return l
    
```

```python
movies['cast']=movies['cast'].apply(replace)
movies['crew']=movies['crew'].apply(replace)
movies['genres']=movies['genres'].apply(replace)
```

```python
# splitting the data
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# Creating the new column tags
movies['tags'] = movies['cast']+ movies['crew'] + movies['overview']+movies['genres']
```

```python
df = movies[['id','title','tags']]
```

```python
df
```

```python
# joining the words
df['tags'] = df['tags'].apply(lambda x: " ".join (x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer # Converting text to vector
```

```python
# Initializing and removing Stop words
cv = CountVectorizer(stop_words = 'english')
```

```python
# Convert text to vector
vector = cv.fit_transform(df['tags']).toarray()
```

# Model Building

```python
from sklearn.metrics.pairwise import cosine_similarity # Finding similarity between attributes
```

```python
# Finding Cosine similarity between the movies
similarity=cosine_similarity(vector)
```

```python
# Sorting on the basis of cosine similarity
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x : x[1])
```

```python
# Movie Recommendation 
def recomend(movie):
    text = df[df['title']==movie].index[0]
    movie_list = sorted(list(enumerate(similarity[text])),reverse=True , key = lambda x: x[1])
    for i in movie_list[1:11]:
        print (df.iloc[i[0]].title)
        
    
    
```

```python
recomend('John Carter')
```
